Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a popular and lightweight web framework written in Python. It is designed to be simple, flexible, and easy to use, making it a great choice for developing webapplication and API and other related project.

Advantages of Flask Framework:-

1. It is light weight framework that offers hassle-free development.
2. it is suitable for small project.
3. Offers a built-in development server and fast debugger.
4. Easily scalable for the application.
5. support for secure cookies.
6. provide flexibilty for developer to experiment with their modules and architecture.
